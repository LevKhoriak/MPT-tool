My inequality constraint is that the total portfolio return must be not less than a given return R

```python
def portfolio_return(w1, w2):
    return w1 * return1 + w2 * return2
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

delta = 0.01
w1 = np.arange(0, 1, delta)
w2 = np.arange(0, 1, delta)

W1, W2 = np.meshgrid(w1, w2)
V = portfolio_variance(W1, W2)
Ret = portfolio_return(W1, W2)

fig, ax = plt.subplots(1, 2)
fig.set_size_inches(15, 5)
PV = ax[0].contour(W1, W2, V)
ax[0].clabel(PV, fontsize=10)
ax[0].set_title('Test')
PR = ax[1].contour(W1, W2, Ret)
ax[1].clabel(PR, fontsize=10)
plt.show()
```
